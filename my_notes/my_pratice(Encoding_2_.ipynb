# Encoding---->
### means changing the categorical data to the numerical data
#### Label ----> changing one column at a time
#### Ordinal ----> taking all the subcategories of all columns together than changing it
#### OneHot ----> will divide the data subcategories into dummy columns but does not give them name
#### Get Dummies----> will divide the data subcategories into dummy columns also give them name.
