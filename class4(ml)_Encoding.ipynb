# Encoding---->
Transforming your categorical data to the numerical data.

(1).Label ---> It will encode data 1-1 columns

(2).Ordinal-----> It will encode the data more than 1 columns

(3).OneHot---> It will create new columns based on sub-categories but it has only values noy names.

(4).get_dummies----> It will create new columns based on sub-categories but it shows column name with values

## (1).LableEncoder---->

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/toyota - toyota.csv")
```

```python
x=df.drop(columns=['price'])
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
x['model']=lb.fit_transform(x['model'])
x['fuelType']=lb.fit_transform(x['fuelType'])
x['transmission']=lb.fit_transform(x['transmission'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_sc=sc.fit_transform(x_train)
```

```python
x_train_new=pd.DataFrame(x_train_sc,columns=x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```

## OrdinalEncoding

```python
df=pd.read_csv("/content/covid_toy - covid_toy.csv")
df.head(3)
```

```python
df=df.dropna()
```

```python
df=df.drop(columns=['age','fever'])
```

```python
df['city'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df.head(2)
```

```python
oe=OrdinalEncoder(categories=[['Male','Female'],
 ['Mild','Strong'],
  ['Kolkata','Bangalore','Delhi','Mumbai'],
   ['Yes','No']])
```

```python
df_ar=oe.fit_transform(df)
```

```python
df_new=pd.DataFrame(df_ar,columns=df.columns)
```
