# FUNCTION TRANSFORMERS---->
### The function transformers is a toll in sckit-learn a popular pyhton library in ml that allows you to apply a specified function to the input data. the function transformers can be usefull for performing custom transformation of imput data in ml pipeline.
##### function transformer takes as input a single function that will be applied to each sample in the data .the finction can not be any python function that takes a single argument,such as lambda function or a user defined function,the function should return the transformed sample.

```python
from sklearn.preprocessing  import FunctionTransformer
import numpy as np
```

```python
x=np.array([[1,2],[3,4]])

# define transformation fucntion

log_transform=FunctionTransformer(np.log1p)

# apply transformation to the data

x_transformed=log_transform.transform(x)

# view the transformed data

print(x_transformed)
```

```python
# practical usecase

# 1. Custom Feature Engineering
x=np.array([[1,2],[3,4]])
# define the custom feature engineering function
def squ(x):
    return np.hstack((x,x**2))

square_transform=FunctionTransformer(squ)

x_transformed=square_transform.transform(x)

print(x_transformed)
```

```python
x=np.array([[1,2],[3,4]])

def my_scalling(x):
    return x/np.max(x)

custom_transformer=FunctionTransformer(my_scalling)

x_transformed=custom_transformer.transform(x)

print(x_transformed)
```

```python
x=np.array([[1,2],[3,np.nan]])

def my_cleaning(x):
    x[np.isnan(x)]=0
    return x

custom_transform=FunctionTransformer(my_cleaning)

x_transformer=custom_transform.transform(x)

print(x_transformer)
```
